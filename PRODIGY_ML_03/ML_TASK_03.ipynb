```python
input_dir = r"D:\Prodigy Intern\dogs-vs-cats\train"
data= []
labels =[]
categories =['cat','dog']
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir,category)):
        img_path = os.path.join(input_dir, category, file)
        print(img_path)
        img = imread(img_path)
        img_resize = resize(img ,(15,15))
        data.append(img_resize.flatten())
        labels.append(category_idx)
```

```python
data = np.array(data)
```

```python
labels=np.array(labels)
```

```python
data.shape
```

```python
labels.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,labels,test_size=0.2,random_state =42)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
img_path = r"D:\Prodigy Intern\dogs-vs-cats\test1\92.jpg"
img = plt.imshow(imread(img_path))
image = imread(img_path)
resize_img = resize(image , (15,15))
final_img = resize_img.flatten()
final_img = np.array(final_img)
final_image = final_img.reshape(1,-1)
pred = model.predict(final_image)
fin_pred = pred[0]
if(fin_pred == 0):
    print("Cat")
else:
    print("Dog")
```
